```python

plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=30, kde=True)
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the relationship between median income and house value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
#Plot the relationship between median income and house value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Prepare the data
X = df[['median_income']]
y = df['median_house_value']

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict house values
predictions = model.predict(X)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(df['median_income'], df['median_house_value'], label='Actual Values')
plt.plot(df['median_income'], predictions, color='red', label='Predicted Values')
plt.title('Linear Regression: Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.show()
```
